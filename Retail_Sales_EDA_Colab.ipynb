**Exploratory Data Analysis (EDA) on Retail Sales Data**


In this project, I worked with a dataset containing information about retail sales. The goal is
to perform exploratory data analysis (EDA) to uncover patterns, trends, and insights that can
help the retail business make informed decisions.

## **Data loading**



```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/menu.csv')

# Display the first 5 rows
display(df.head())

# Display the data types
display(df.info())
```

**Findings:**
- Each row corresponds to a food item from the McDonald’s menu.
- Columns include both categorical (`Category`, `Item`) and numerical data (`Calories`, `Total Fat`, `Sodium`, etc.).
- Serving sizes are recorded in ounces and grams.


## **Data cleaning**



```python
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values)
```

**Findings:**
- `Category`, `Item`, and `Serving Size` are object (string) types.
- Nutritional attributes like `Calories`, `Protein`, and `Sodium` are integers/floats.
- Dataset is clean with no null values, so preprocessing will be minimal.


```python
display(df.info())
display(df.describe())

import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(4, 6, i + 1)
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(4, 6, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

## **Descriptive statistics**



```python
descriptive_stats = df.describe()
display(descriptive_stats)
```

Interpret the calculated descriptive statistics for key numerical columns to understand their central tendency, spread, and range.



```python
print("Interpretation of Descriptive Statistics for Key Nutritional Columns:")
print("\nCalories:")
print(f"  - Mean: {descriptive_stats.loc['mean', 'Calories']:.2f}")
print(f"  - Median: {descriptive_stats.loc['50%', 'Calories']:.2f}")
print(f"  - Standard Deviation: {descriptive_stats.loc['std', 'Calories']:.2f}")
print(f"  - Range: {descriptive_stats.loc['max', 'Calories'] - descriptive_stats.loc['min', 'Calories']:.2f}")
print("  Interpretation: The average calorie count is around 368, with a wide range from 0 to 1880, indicating significant variability in calorie content across menu items.")

print("\nTotal Fat:")
print(f"  - Mean: {descriptive_stats.loc['mean', 'Total Fat']:.2f}")
print(f"  - Median: {descriptive_stats.loc['50%', 'Total Fat']:.2f}")
print(f"  - Standard Deviation: {descriptive_stats.loc['std', 'Total Fat']:.2f}")
print(f"  - Range: {descriptive_stats.loc['max', 'Total Fat'] - descriptive_stats.loc['min', 'Total Fat']:.2f}")
print("  Interpretation: The average total fat is about 14g, with a large standard deviation relative to the mean, suggesting a varied fat content. The range is also quite wide, from 0g to 118g.")

print("\nSugars:")
print(f"  - Mean: {descriptive_stats.loc['mean', 'Sugars']:.2f}")
print(f"  - Median: {descriptive_stats.loc['50%', 'Sugars']:.2f}")
print(f"  - Standard Deviation: {descriptive_stats.loc['std', 'Sugars']:.2f}")
print(f"  - Range: {descriptive_stats.loc['max', 'Sugars'] - descriptive_stats.loc['min', 'Sugars']:.2f}")
print("  Interpretation: On average, items contain about 29g of sugar. The median is lower than the mean, which could indicate a skew towards higher sugar content in some items. The range is from 0g to 128g, highlighting a considerable variation in sugar levels.")

print("\nProtein:")
print(f"  - Mean: {descriptive_stats.loc['mean', 'Protein']:.2f}")
print(f"  - Median: {descriptive_stats.loc['50%', 'Protein']:.2f}")
print(f"  - Standard Deviation: {descriptive_stats.loc['std', 'Protein']:.2f}")
print(f"  - Range: {descriptive_stats.loc['max', 'Protein'] - descriptive_stats.loc['min', 'Protein']:.2f}")
print("  Interpretation: The average protein content is around 13g. The distribution is somewhat spread out, with a range from 0g to 87g, showing variety in protein levels.")
```

**Findings:**
- **Calories**: Mean ~ 368, with items ranging from very low-calorie (0–50) to extremely high (~1100+).  
- **Total Fat**: Averages around 20g, with some items containing 50g+.  
- **Sodium**: High variability; many items exceed the recommended daily intake in a single serving.  
- **Protein**: Ranges widely — some items are protein-rich (30g+) while others have very little.  

This highlights how menu items differ significantly in nutritional content.

## **Customer and product analysis**



```python
# Group by 'Category' and calculate the mean of numerical columns
category_performance = df.groupby('Category').mean(numeric_only=True)

# Display the category performance DataFrame
display(category_performance)

# Sort by 'Calories' in descending order and display
category_performance_sorted_calories = category_performance.sort_values(by='Calories', ascending=False)
display(category_performance_sorted_calories)
```

```python
# Identify the top 10 items based on 'Calories'
top_10_calories = df.sort_values(by='Calories', ascending=False).head(10)
print("Top 10 items by Calories:")
display(top_10_calories)

# Identify the bottom 10 items based on 'Calories'
bottom_10_calories = df.sort_values(by='Calories', ascending=True).head(10)
print("\nBottom 10 items by Calories:")
display(bottom_10_calories)
```

**Findings:**
- **Breakfast items**: Moderate calories, balanced protein, but high in fat.  
- **Beverages**: Extremely high in sugars (some exceeding 80g).  
- **Desserts**: High sugar + fat content, with minimal protein.  
- **Salads**: Lower calorie options, often higher in fiber and vitamins.  

This analysis helps identify which categories are “healthier” vs. “less healthy.”


## **Visualization**



I have created bar plots to visualize the average calorie count per category, the top 10 items by calories, and the bottom 10 items by calories. Also, created a scatter plot to show the relationship between Calories and Total Fat.



```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Bar plot of average calorie count per category
plt.figure(figsize=(12, 6))
sns.barplot(x=category_performance_sorted_calories.index, y='Calories', data=category_performance_sorted_calories, hue=category_performance_sorted_calories.index, palette='viridis', legend=False)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Category')
plt.ylabel('Average Calories')
plt.title('Average Calorie Count per Category')
plt.tight_layout()
plt.show()
```

## Findings: Average Calorie Count per Category

The bar chart above represents the **average calorie count across different food categories**. Key observations are as follows:

1. **Highest Calorie Categories**  
   - *Chicken & Fish* has the highest average calorie count (~560 calories).  
   - *Smoothies & Shakes* and *Breakfast items* also rank high, each averaging above 520 calories.  
   - These categories contribute significantly to higher calorie intake.

2. **Moderate Calorie Categories**  
   - *Beef & Pork* falls just below 500 calories on average.  
   - *Coffee & Tea* (~ 285 calories) and *Salads* (~ 270 calories) show moderate calorie levels compared to heavy meals.  

3. **Lowest Calorie Categories**  
   - *Snacks & Sides* (~ 245 calories) and *Desserts* (~ 225 calories) provide relatively lighter options.  
   - *Beverages* have the lowest average calorie count (~ 115 calories), making them the least calorie-dense category.

### Insights
- **Main meals (Chicken, Fish, Breakfast, Smoothies & Shakes)** are calorie-heavy, which can quickly add up in a diet.  
- **Light options (Beverages, Salads, Snacks & Sides)** may be preferable for calorie-conscious individuals.  
- There’s a **significant gap (~450 calories)** between the highest (Chicken & Fish) and lowest (Beverages) average calorie categories.  

This analysis highlights the importance of **food category choices** in managing overall calorie intake.


```python
# 2. Bar plot of top 10 items by calorie count
plt.figure(figsize=(12, 6))
sns.barplot(x='Item', y='Calories', data=top_10_calories, palette='plasma')
plt.xticks(rotation=90, ha='right')
plt.xlabel('Item')
plt.ylabel('Calories')
plt.title('Top 10 Items by Calorie Count')
plt.tight_layout()
plt.show()

# Filter out items with 0 calories to show meaningful low-calorie items
df_filtered_calories = df[df['Calories'] > 0]

# Identify the bottom 10 items by calorie count from the filtered data
bottom_10_calories_filtered = df_filtered_calories.sort_values(by='Calories', ascending=True).head(10)
```

```python
# 3. Bar plot of bottom 10 items by calorie count (excluding zero-calorie items)
plt.figure(figsize=(12, 6))
sns.barplot(x='Item', y='Calories', data=bottom_10_calories_filtered, palette='mako', hue='Item', legend=False)
plt.xticks(rotation=90, ha='right')
plt.xlabel('Item')
plt.ylabel('Calories')
plt.title('Bottom 10 Items by Calorie Count (Excluding Zero-Calorie Items)')
plt.tight_layout()
plt.show()
```

```python

# 4. Scatter plot of Calories vs Total Fat
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total Fat', y='Calories', data=df, alpha=0.6)
plt.xlabel('Total Fat (g)')
plt.ylabel('Calories')
plt.title('Relationship between Calories and Total Fat')
plt.grid(True)
plt.tight_layout()
plt.show()
```

1. **Bar Charts** – Compare average calories, sugar, and fat per category.  
2. **Line Plots** – Track nutrient variations across menu items.  
3. **Heatmaps** – Show correlation among nutritional variables.

**Findings:**
- Strong positive correlation between **Calories** and **Total Fat / Sugars**.  
- Categories like **Beverages & Desserts** dominate sugar contribution.  
- **High-calorie items** usually have high sodium and fat levels as well.  


## **Recommendations**




Based on the EDA, we can make actionable suggestions:

1. **For Customers:**
   - Choose **salads and grilled items** for lower calorie intake.
   - Limit **beverages and desserts**, as they are sugar-heavy.
   - Pay attention to sodium intake, especially in burgers and sandwiches.

2. **For McDonald’s (Business Insight):**
   - Promote **low-calorie, high-protein options** for health-conscious customers.
   - Introduce **reduced-sodium versions** of popular items.
   - Highlight menu categories with **balanced nutrition** in marketing.

These recommendations can guide both consumers (healthy choices) and the business (product positioning).
